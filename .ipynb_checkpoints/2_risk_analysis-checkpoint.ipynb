```python
import utils
import risk_model

from bs4 import BeautifulSoup
import requests
import re
from pprint import pprint
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (28, 16)
```

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = BeautifulSoup(r.text,'html.parser')
T=[]
tickers = soup.find_all('a',{'class':'external text'},{'rel':'nofollow'})
for i in tickers[:-180]:
    if i.text == 'reports':
        del i
    else:
        T.append(i.text)
sp500 = T[:-11]
```

```python
universe = sp500[:40]
```

```python
#close.to_csv('close.csv')
```

```python
close = pd.read_csv('close.csv', index_col = 'Date')
```

```python
close.index = pd.to_datetime(close.index, format='%Y%m%d')
```

```python
close
```

```python
returns  = close.apply(lambda x:(x - x.shift(1))/x).iloc[1:,:].fillna(0)
```

```python
num_factor_exposures = 10
pca = risk_model.fit_pca(close=close,num_factor_exposures=num_factor_exposures,svd_solver='full')
```

```python
plt.title('Explained Variance')
plt.bar(np.arange(num_factor_exposures), pca.explained_variance_ratio_);
```

```python
pca.components_.shape[0]
```

```python
Risk_Model = {}
Risk_Model['factor_betas'] = risk_model.factor_betas(pca, returns)
```

```python
Risk_Model['factor_betas'].tail()
```

```python
Risk_Model['factor_returns'] = risk_model.factor_returns(
    pca,
    returns,
    returns.index,
    np.arange(num_factor_exposures))

Risk_Model['factor_returns'].cumsum().plot(legend=None);
```

```python
ann_factor = 252
Risk_Model['factor_cov_matrix'] = risk_model.factor_cov_matrix(Risk_Model['factor_returns'], ann_factor)

Risk_Model['factor_cov_matrix']
```

```python
Risk_Model['idiosyncratic_var_matrix'] = risk_model.idiosyncratic_var_matrix(returns, Risk_Model['factor_returns'], risk_model['factor_betas'], ann_factor)

Risk_Model['idiosyncratic_var_matrix']
```

```python
Risk_Model['idiosyncratic_var_vector'] = risk_model.idiosyncratic_var_vector(returns, Risk_Model['idiosyncratic_var_matrix'])

Risk_Model['idiosyncratic_var_vector']
```

- $ X $ is the portfolio weights
- $ B $ is the factor betas
- $ F $ is the factor covariance matrix
- $ S $ is the idiosyncratic variance matrix
