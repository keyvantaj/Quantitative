```python
annual_cleaned = utils.cleaning_multiindex_dataframe(annual, 20)
quarter_cleaned = utils.cleaning_multiindex_dataframe(quarter, 20)
```

```python
fig, axs = plt.subplots(8,9, figsize = (20,20),sharex=True)
axs = axs.flatten()
x = annual_cleaned.index
plt.title('Annual Overview of Financial Statements for {}'.format(symbol))
for i,name in list(zip(range(len(axs)),annual_cleaned.columns[:len(axs)])):
    axs[i].set_title(name,fontsize=8)
    axs[i].plot(x,annual_cleaned[name].values)
    for tick in axs[i].get_xticklabels():
        tick.set_rotation(45)
        tick.set_fontsize(10)
    for tick in axs[i].get_yticklabels():
        tick.set_rotation(45)
        tick.set_fontsize(8)
plt.show()
```

```python
fig, axs = plt.subplots(8,7, figsize = (20,20),sharex=True)
axs = axs.flatten()
x = quarter_cleaned.index
plt.title('Quarter Overview of Financial Statements for  {}'.format(symbol))
for i,name in list(zip(range(len(axs)),quarter_cleaned.columns[:len(axs)])):
    axs[i].set_title(name,fontsize=8)
    axs[i].plot(x,quarter_cleaned[name].values)
    for tick in axs[i].get_xticklabels():
        tick.set_rotation(45)
        tick.set_fontsize(10)
    for tick in axs[i].get_yticklabels():
        tick.set_rotation(45)
        tick.set_fontsize(8)
plt.show()
```

```python
x = quarter.index
plt.figure(figsize=(40,20))
plt.plot(x,quarter[['Net Income']].values)
plt.xticks(rotation=70,fontsize=30)
plt.yticks(rotation=70,fontsize=30)
plt.grid(True)
```
